The goal of this model will be to predict the minutes a show or movie was watched for each viewing period from the Netflix profile 'Brette'. For the analysis, after running the data through a PyCaret regression model that evaluated the data and chose a model that would best predict the outcomes, we will be using a Gradient Boosting Regressor model.Below is a link to the Y-Data Report of this dataset:

## Importing Libraries

##### Below are the libraries I will use to complete my regression model.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy as sp
from pycaret.regression import *


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error,  r2_score, explained_variance_score
```

## Loading Netflix Dataset

```python
df = pd.read_csv('Netflix.csv')
df.head()
```

## Filtering & Cleaning the Data

##### I will first look for nulls in the dataset before to make sure there are no empty rows for the analysis

```python
df.isnull().sum()
```

##### Then I will look for the data types to see if any need to be changed

```python
df.info()
```

##### Changing the text and Date columns to category fields from objects:

```python
df[['Profile Name','Title','Type','Genre','Device Type']] = df[['Profile Name','Title','Type','Genre','Device Type']].astype('category')
```

```python
df['Date'] = pd.to_datetime(df['Date']) #passing through the 'datetime' type
```

### New Columns

#### Before beginning the analysis, for the model to perform better I will be breaking out the 'Date' column in to 'Year', 'Month', and 'Day', and converting the 'Duration_Seconds' to a new column 'Minutes'. This way the dataframe will be able to predict better scores and can be better visually represeted.

##### Altering the 'Date' column so that the year, month, and day are separated and moved to the beginning of the dataframe

```python
df['Year'] = df['Date'].dt.year #taking the year from the 'Date' column and creating a column of it's own
df['Month'] = df['Date'].dt.month #taking the month from the 'Date' column and creating a column of it's own
df['Day'] = df['Date'].dt.day #taking the day from the 'Date' column and creating a column of it's own
```

```python
cols = ['Year', 'Month', 'Day'] + [col for col in df.columns if col not in ['Year', 'Month', 'Day']] #telling the df to place the new columns at the 
df = df[cols]                                                                                  
```

```python
df = df.drop(columns = 'Date') # dropping the column 'Date' once we have the new columns
```

##### Converting 'Duration_Seconds' to 'Minutes'

```python
df['Minutes'] = df['Duration Seconds'] / 60 #dividing the total seconds by 60 to get back a new column with the total minutes
df = df.drop(columns='Duration Seconds') #dropping the 'Duration Seconds' column
```

##### The last step to cleaning the data is dropping the 'Title' column so the Gradient Boosting Regressor model will be able to read the dataframe more easily

```python
df = df.drop(columns='Title') #Dropping the column 'Title' from the dataframe and passing through to the dataframe
df
```

##### Now that the new columns are created and the dtypes are properly set, I will filter the dataframe so that it is just data from my Netflix account

The below code is passing a filtered dataframe with only rows that contain the profile name 'Brette' to the original dataframe

```python
df = df[df['Profile Name'] == 'Brette']
```

### Dataframe Features and Definitions

##### Now that the dataframe is ready below are the definitions for each features and their datatypes:

##### Year : The year the show or movie was watched by my Netflix account, datatype: int32
##### Month: The month the show or movie was watched by my Netflix account, datatype: int32
##### Day : The day the show or movie was watched by my Netflix account, datatype: int32
##### Profile Name: The name of the profile that watched Netflix content, in this dataframe there is only 'Brette', datatype: category
##### Type : Whether the content watched was either a 'Show' or 'Movie', datatype: category
##### Genre: The category that the movie or show is under on Netflix, datatype: category
##### Device Type: The device that the movie or show was streamed on while watching, datatype: category
##### Minutes: The length of time that the show or movie was watched in one sitting period, datatype: float64

```python
df
```

### Visualizing the Data

```python
# Summarize the data
df_summed_genre = df.groupby('Genre', observed=True)['Minutes'].sum().reset_index() #Grouping the genres by the total minutes watched
df_summed_type = df.groupby('Type', observed=True)['Minutes'].sum().reset_index() #Grouping the types movie and show by the total minutes watched
df_summed_devices = df.groupby('Device Type', observed=True)['Minutes'].sum().reset_index() #Grouping the devices used by total minutes watched
df_summed_years = df.groupby('Year', observed=True)['Minutes'].sum().reset_index() #Grouping each year by the total minutes watched
top_3_devices = df_summed_devices.nlargest(3, 'Minutes') #Taking the group by above and filtering it to show the top 3
```

## Beginning Analysis

##### To begin the analysis, I will split the dataframe in to X and y to define the target for the train test split below

```python
X = df.drop(columns=df[['Minutes']],axis=1) #dropping the 'Minutes' column from the dataframe for the y variable
X #X is a 2D feature shown as a pandas dataframe
```

```python
y = df['Minutes'] #passing through the 'Minutes' column to the y variable
y # the target, which is now a single feature
```

##### Now before beginning the train test split, I will pass X through a one hot encoder since the dataframe includes many categories

```python
X = pd.get_dummies(X, drop_first=True) #Analyzing True or False for each row in the dataframe, showing if it is present in the X dataframe
X
```

## Train Test Split

##### The X and y data is now ready to be passed through the train test split to ensure the data is not overloaded and our machine learning algorithm can better predict in the model

X_train and X_test will remain 2D and y_train and y_test will be taken from the 1D column we created above
For this train test split, we will be using a test size of 30% vs 70% and a random state of 42

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

##### Now that the train variables are set, we can pass the 2D X variables through a scaler that will help normalize the data in the Gradient Boosting regression model. I have chosen the MinMaxScaler since there is a lot of variability in the dataset with many outliers.

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train, X_test # a visual of the X_train and X_test arrays after being scaled and transformed
```

## GBR Regression Model

##### Below I am defining the model as the Gradient Boosting Regressor Machine Learning module
##### The Gradient Boosting Regressor model attempts to prevent over-fitting the dataframe since there is a wide variety of variables and predictons in the dataframe. Since there are not many correlations in the data and it is assumed it will not be normalized, this model also performs well with non-linear relationships.

```python
mdl = GradientBoostingRegressor(random_state=42)
mdl #Using the exact model the PyCaret regression model gave
```

```python
mdl.fit(X_train,y_train) #Fitting the train test split we did above to the GBR machine learning model
```

##### Now we can create the predictions from the model, we can pass through the X_test variable so we will be able to compare it to the y_test array that are the target values

```python
y_hat = mdl.predict(X_test) #passing the predictions for X_test into y_hat for the scores we will calculate
y_hat
```

```python
len(y_test), len(y_hat) #double checking to see that the lengths of both y_hat and y_test match since it is necessary for the analysis
```

##### Now that the model is fitted, we can begin to analyze the data, below are calculations for the mean, median, and standard deviation of the dataset

```python
df['Minutes'].mean(), df['Minutes'].median(), df['Minutes'].std()
```

##### Creating the minimum, maximum and range variables for our scores

```python
ymin = np.min(y_test) #finding the min max and range of y_test (knowing the base units)
ymax = np.max(y_test)
yrange = (np.max(y_test) - np.min(y_test))
# Establish the base units to compare the metrics  ....
ymin, ymax, yrange
```

#### Mean absolute error score for the GBR model

```python
mae = mean_absolute_error(y_test,y_hat)
print(f'An MAE of {mae:.3} means the avg prediction error is about {mae / yrange * 100:.3}% of the total range of y_test.')
```

#### Mean squared error score for the GBR model

```python
mse = mean_squared_error(y_test, y_hat) #the mean squared error that is reported in square units that we will use for the rmse calculation
print(f'The mse is {mse}')
```

#### Square root of the mean squared error score for the GBR model

```python
rmse = np.sqrt(mean_squared_error(y_test, y_hat))
print(f'An RMSE of {rmse:.3} means the avg prediction error is about {rmse / yrange * 100:.3}% of the total range of y_test.')
```

#### The r-squared score of the GBR model

```python
rsq = r2_score(y_test, y_hat) #0.14871
print(f'An r-square score of {rsq:.5} means approx {rsq * 100:.3}% in the actual (y_test) values is explained by the predicted (y-hat) values.\nThe remaining {(1-rsq) * 100:.3} of the variance is due to unknown factors (i.e. factors not captured by the model). ') 
```

#### Adjusted score of the GBR model

```python
r_squ = r2_score(y_test, y_hat) #passing the r-squared percentage in that we just created above
n = len(y) #the number of objects in the y variable
k = X.shape[1] #the shape of x that is the number of predictors
adj_rsq = 1 - (1 - r_squ) * (n - 1) / (n - k - 1)
adj_rsq
```

#### Creating a dataframe to view how the model predicted against the actual results in the original dataframe

```python
yhat_df = pd.DataFrame({'Actuals':y_test, 'Preds': y_hat})
yhat_df['Diff'] = np.abs(yhat_df['Actuals'] - yhat_df['Preds']) 
yhat_df['Residuals'] = yhat_df['Actuals'] - yhat_df['Preds']
yhat_df
```

##### Analyzing the mean of the dataframe we created above

```python
yhat_df.mean() 
```
